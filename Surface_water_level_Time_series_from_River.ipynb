```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/sample_data/filr/water-level-heights-Rivers-State.csv'
data = pd.read_csv(file_path, sep="\t", skiprows=2)

# Rename columns for better clarity
data.columns = ["Date", "Water Level", "Water Level Error"]

# Convert the 'Date' column to datetime format and drop rows with invalid dates or missing values
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['Water Level'] = pd.to_numeric(data['Water Level'], errors='coerce')
data.dropna(subset=['Date', 'Water Level'], inplace=True)

# Extract year and month
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month

# Filter data for the years 2018 to 2024
filtered_data = data[(data['Year'] >= 2018) & (data['Year'] <= 2024)]

# Group by year and month to calculate the average water level
monthly_avg_water_levels = filtered_data.groupby(['Year', 'Month'])['Water Level'].mean().unstack()

# Plot the average water levels
plt.figure(figsize=(12, 8))
for year in monthly_avg_water_levels.index:
    plt.plot(monthly_avg_water_levels.columns, monthly_avg_water_levels.loc[year], label=str(year))

# Add labels, title, and legend
plt.title('Yearly Water Level Trends (2018-2024) Rivers State', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Water Level (m)', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year', fontsize=10)
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
